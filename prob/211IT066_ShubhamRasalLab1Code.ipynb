```python

import pandas as pd
import random
import math
import matplotlib.pyplot as plt
```

```python

# Creating a dataset
# 1. Create a dataset of 1000 rows and 3 columns.
# 2. First column is patient ID (S.no).
# 3. Second column is test result (0 or 1).
# 4. Third column is actual disease status (0 or 1).

# Creating a dataset of 1000 rows and 3 columns
dataset = pd.DataFrame(columns=['Patient ID', 'Test Result', 'Actual Disease Status'])

# Creating a list of 1000 patient IDs
patient_id = []
for i in range(1, 1001):
    patient_id.append(i)

# Creating a list of 1000 test results
test_result = []
for i in range(0, 1000):
    test_result.append(random.randint(0, 1))

# Creating a list of 1000 actual disease status
actual_disease_status = []
for i in range(0, 1000):
    actual_disease_status.append(random.randint(0, 1))

# Adding the lists to the dataset
dataset['Patient ID'] = patient_id
dataset['Test Result'] = test_result
dataset['Actual Disease Status'] = actual_disease_status

# Printing the dataset
print(dataset)


```

```python

# Calculating the probability of having the disease
# 1. Calculate the probability of having the disease.

# Calculating the probability of having the disease

probability_of_having_the_disease = dataset['Actual Disease Status'].sum() / len(dataset['Actual Disease Status'])

# Printing the probability of having the disease
print("Probability of having the disease: ", probability_of_having_the_disease)

```

```python

#2. Calculate the probability of getting a positive test result.

# Calculating the probability of getting a positive test result
probability_of_getting_a_positive_test_result = dataset['Test Result'].sum() / len(dataset['Test Result'])

# Printing the probability of getting a positive test result
print("Probability of getting a positive test result: ", probability_of_getting_a_positive_test_result)
```

```python

# 3. Calculate the probability of getting a positive test result given that a patient has the disease.

# Calculating the probability of getting a positive test result given that a patient has the disease
positive_test_result_given_patient_has_disease = (dataset['Test Result'] & dataset['Actual Disease Status']).sum() / dataset['Actual Disease Status'].sum()

# Printing the probability of getting a positive test result given that a patient has the disease
print("Probability of getting a positive test result given that a patient has the disease: ", positive_test_result_given_patient_has_disease)
```

```python

# 4. Calculate the probability of having the disease given a positive test result using Bayes’ theorem.

# Calculating the probability of having the disease given a positive test result using Bayes’ theorem
probability_of_having_the_disease_given_positive_test_result = (positive_test_result_given_patient_has_disease * probability_of_having_the_disease) / probability_of_getting_a_positive_test_result

# Printing the probability of having the disease given a positive test result using Bayes’ theorem
print("Probability of having the disease given a positive test result using Bayes’ theorem: ", probability_of_having_the_disease_given_positive_test_result)
```

```python

# 5. Calculate confusion matrix, accuracy, sensitivity and specificity.

# Calculating confusion matrix, accuracy, sensitivity and specificity
confusion_matrix = [[0, 0], [0, 0]]

for i in range(0, len(dataset['Test Result'])):
    if dataset['Test Result'][i] == 1 and dataset['Actual Disease Status'][i] == 1:
        confusion_matrix[0][0] += 1
    elif dataset['Test Result'][i] == 1 and dataset['Actual Disease Status'][i] == 0:
        confusion_matrix[0][1] += 1
    elif dataset['Test Result'][i] == 0 and dataset['Actual Disease Status'][i] == 1:
        confusion_matrix[1][0] += 1
    elif dataset['Test Result'][i] == 0 and dataset['Actual Disease Status'][i] == 0:
        confusion_matrix[1][1] += 1

accuracy = (confusion_matrix[0][0] + confusion_matrix[1][1]) / (confusion_matrix[0][0] + confusion_matrix[0][1] + confusion_matrix[1][0] + confusion_matrix[1][1])

sensitivity = confusion_matrix[0][0] / (confusion_matrix[0][0] + confusion_matrix[1][0])

specificity = confusion_matrix[1][1] / (confusion_matrix[1][1] + confusion_matrix[0][1])



# Printing confusion matrix, accuracy, sensitivity and specificity
print("Confusion Matrix: \n", confusion_matrix)

print("Accuracy: ", accuracy)

print("Sensitivity: ", sensitivity)

print("Specificity: ", specificity)


```

```python

# 6. Generate classification reports including precision, recall and F1-score.

# Generating classification reports including precision, recall and F1-score
classification_report = pd.DataFrame(columns=['Precision', 'Recall', 'F1-Score'])

classification_report['Precision'] = [confusion_matrix[0][0] / (confusion_matrix[0][0] + confusion_matrix[0][1]), confusion_matrix[1][1] / (confusion_matrix[1][1] + confusion_matrix[1][0])]

classification_report['Recall'] = [confusion_matrix[0][0] / (confusion_matrix[0][0] + confusion_matrix[1][0]), confusion_matrix[1][1] / (confusion_matrix[1][1] + confusion_matrix[0][1])]

classification_report['F1-Score'] = [2 * ((classification_report['Precision'][0] * classification_report['Recall'][0]) / (classification_report['Precision'][0] + classification_report['Recall'][0])), 2 * ((classification_report['Precision'][1] * classification_report['Recall'][1]) / (classification_report['Precision'][1] + classification_report['Recall'][1]))]



# Printing classification reports including precision, recall and F1-score
print("Classification Report: \n", classification_report)
```
